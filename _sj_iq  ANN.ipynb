```
sj_model = Sequential()

# The Input Layer :
sj_model.add(Dense(128, kernel_initializer='normal',input_dim = sj_train_features.shape[1], activation='relu'))

# The Hidden Layers :
sj_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
sj_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
sj_model.add(Dense(256, kernel_initializer='normal',activation='relu'))

# The Output Layer :
sj_model.add(Dense(1, kernel_initializer='normal',activation='linear'))

# Compile the network :
sj_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
sj_model.summary()
```
