```python
training_set = pd.read_csv('dengue_features_train.csv')
```

```python
training_labels = pd.read_csv("dengue_labels_train.csv")
```

```python
training_set.shape
```

```python
training_labels = training_labels.iloc[:,3].values
```

```python
training_set['total_cases'] = training_labels
```

```python
training_set
```

```python
train = training_set
```

```python
months = []
for date in train['week_start_date']:
    x = date.split("-")
    months.append(int(x[1]))
```

```python
train['week_start_date'] = months
```

```python
train.fillna(method="ffill", inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
train = scaler.fit_transform(train)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
train = pca.fit_transform(train)
```

```python
total = sum(pca.explained_variance_)
k = 0
cur_var = 0 
while cur_var / total < 0.999:
    cur_var += pca.explained_variance_[k]
    k = k+1
k
```
